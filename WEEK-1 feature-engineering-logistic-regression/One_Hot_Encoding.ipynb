## one hot encoding using Scikit learn

```python

ohe = OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
```

```python
X_train_new =ohe.fit_transform(X_train[['fuel' , 'owner']])
X_test_new =ohe.transform(X_test[['fuel' , 'owner']])
```

```python
counts = df['brand'].value_counts()
```

```python
df['brand'].nunique
threshold = 100
```

```python
counts[counts <= threshold].index
```

```python
repl = counts[counts <= threshold].index
```

```python
brand_ohe = pd.get_dummies(X_train['brand'].replace(repl, 'uncommon'), dtype=np.int32)
```

```python
fuel_owner_ohe = pd.DataFrame(X_train_new,columns=ohe.get_feature_names_out(['fuel', 'owner']),index=X_train.index
)
```

```python
remaining = X_train[['km_driven']]
```

```python
X_final = pd.concat([
    X_train[['km_driven']].reset_index(drop=True),
    brand_ohe.reset_index(drop=True),
    fuel_owner_ohe.reset_index(drop=True)
], axis=1)
```

```python
fuel_owner_ohe = fuel_owner_ohe.reset_index(drop=True)
brand_ohe = brand_ohe.reset_index(drop=True)
remaining = X_train[['km_driven']].reset_index(drop=True)
```

```python
X_final = pd.concat(
    [remaining, brand_ohe, fuel_owner_ohe],
    axis=1
)
```
